```python
def check_dupl(ele):
    ''' Check if given list contains any duplicates '''
    if len(ele) == len(set(ele)):
        return False
    else:
        return True
result = check_dupl([1,2,3,4,3,2,1])
if result:
    print("Yes, list contains duplicates")
else:
    print("no duplicates found in list")
```
